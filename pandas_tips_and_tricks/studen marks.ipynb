step 1: import libraries

```python
import pandas as pd
from datetime import datetime
```

step 2:Create student marks csv

```python
data = [
    ["Muhammad Qasim", "Math", 85, "2024-01-10"],
    ["Muhammad Qasim", "English", 78, "2024-01-10"],
    ["Muhammad Qasim", "Science", 88, "2024-01-10"],
    
    ["Muhammad Qasim", "Math", 90, "2024-03-15"],
    ["Muhammad Qasim", "English", 82, "2024-03-15"],
    ["Muhammad Qasim", "Science", 91, "2024-03-15"]
]

columns = ["student_name", "subject", "marks", "exam_date"]

df = pd.DataFrame(data, columns=columns)
df.to_csv("student_marks.csv", index=False)

df
```

step 4:Function to calculate Grade

```python
def calculate_grade(marks):
    if marks >= 90:
        return "A+"
    elif marks >= 80:
        return "A"
    elif marks >= 70:
        return "B"
    elif marks >= 60:
        return "C"
    else:
        return "Fail"
```

step5: Apply Grade Function

```python
df["grade"] = df["marks"].apply(calculate_grade)
df
```

step6:separate Exam-1 and Exam-2

```python
df["exam_date"] = pd.to_datetime(df["exam_date"])

exam1 = df[df["exam_date"] == df["exam_date"].min()]
exam2 = df[df["exam_date"] == df["exam_date"].max()]
```

step7: compare Exam result

```python
comparison = exam1.merge(
    exam2,
    on="subject",
    suffixes=("_Exam1", "_Exam2")
)

comparison["marks_difference"] = comparison["marks_Exam2"] - comparison["marks_Exam1"]

comparison
```

step8:Final result Export

```python
final_results = comparison[[
    "student_name_Exam1",
    "subject",
    "marks_Exam1",
    "marks_Exam2",
    "marks_difference"
]]

final_results.rename(columns={"student_name_Exam1": "student_name"}, inplace=True)

final_results.to_csv("final_results.csv", index=False)
final_results
```

step9:Subject summary

```python
subject_summary = df.groupby("subject").agg(
    average_marks=("marks", "mean"),
    highest_marks=("marks", "max"),
    lowest_marks=("marks", "min")
).reset_index()

subject_summary.to_csv("subject_summary.csv", index=False)
subject_summary
```
