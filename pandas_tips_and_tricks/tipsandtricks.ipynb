(03)- how to rename columns?

```python
df = pd.DataFrame({'A col':[1,2,3],'B col':[4,5,6]})
df
```

```python
df.rename(columns = {'A col': 'col_a', 'B col':'col_b'}, inplace=True)
df
```

```python

df.columns = ['col_aa', 'col_bb']
df
```

```python
# to replace any character, string
df.columns = df.columns.str.replace('aa', '_a')
```

```python
# adding prefix to columns
df.add_prefix('baba_')
df
```

```python
# adding suffix to columns
df.add_suffix('_baba')
```

```python
len(df)
```

```python
df.shape
```

```python
kashti_1 = df.sample(frac=0.50, random_state=1)
kashti_1.shape
```

```python
kashti_2 = df.drop(kashti_1.index)
kashti_2.shape
```
