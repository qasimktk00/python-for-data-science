```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import titanic data from seaborn library
titanic = sns.load_dataset('titanic')
```

```python
df = titanic.copy()
```

```python
df.plot(kind='scatter', x='age', y='fare', color='red')
```

```python
df.plot(kind='scatter', x='age', y='fare', color='red')
```

```python
# change colors using hex codes
df.plot(kind='scatter', x='age', y='fare', color='#FF5733')
```

```python
df.head()
```

```python
df.groupby('sex')['fare'].mean().plot(kind='bar')
plt.xlabel('Average fare')
plt.ylabel('Gender')
```

```python
# seaborn plots

# bar plot using seaborn on df
sns.scatterplot(data=df, y='fare', x ='age',hue='class')
```

```python
# make scatter plot via plotly
import plotly.express as px

fig = px.scatter(df, x='age', y='fare', color='class')
fig.show()
```

```python
import plotly.graph_objects as go
import pandas as pd

# Fixed URL - single 'https://'
data = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/api_docs/mt_bruno_elevation.csv')

fig = go.Figure(data=[go.Surface(z=data.values)])

fig.update_layout(title='Mt Bruno Elevation', autosize=False,
                  width=500, height=500,
                  margin=dict(l=65, r=50, b=65, t=90))
fig.show()
```

```python
import plotly.express as px
df = px.data.gapminder()
px.scatter(df, x='gdpPercap', y='lifeExp', animation_frame='year', animation_group='country',
              size='pop', color='continent', hover_name='country',
              log_x=True, size_max=55, range_x=[100,100000], range_y=[25,90])
```
