# Matplotlib: Scatter plot on Iris dataset

```python
import matplotlib.pyplot as plt
iris = sns.load_dataset('iris')
plt.scatter(iris['sepal_length'], iris['sepal_width'], c=iris['species'].astype('category').cat.codes)
plt.xlabel('Sepal Length'); plt.ylabel('Sepal Width'); plt.title('Iris Sepal Dimensions')
plt.show()
```

# Seaborn: Heatmap on flights passengers

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load flights dataset and pivot CORRECTLY with keyword arguments
flights = sns.load_dataset('flights').pivot(index='month', columns='year', values='passengers')

# Create heatmap
sns.heatmap(flights, annot=True, cmap='YlOrRd', fmt='d')

# Add title and display
plt.title('Monthly Passenger Traffic by Year')
plt.tight_layout()
plt.show()
```
